# Import of Packages

```python
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.utils import np_utils
from keras import backend as k
k.set_image_dim_ordering('th')\
```

# Loading image data and text data from Csv is done using Data loaders
    But in our case it is bone by load data

```python
(x_train,y_train),(x_valid,y_valid) = mnist.load_data()
```

# Normalizing and Pre preocessing the data
    In our case we have black and white images but in case of other data sets preprocessing is very much required.

```python
# reshape the size of the train and vaklidation sets to this foprmat(exapmle,channels,width, heaight)
x_train = x_train.reshape(x_train.shape[0],1,28,28)
x_valid = x_valid.reshape(x_valid.shape[0],1,28,28)
# normalising the input from 0-225 to 0-1
x_train = x_train/255
x_valid = x_valid/255
# ONE HOT ENCODING FOR y
y_train = np_utils.to_categorical(y_train)
y_valid = np_utils.to_categorical(y_valid)
num_classes = y_valid.shape[1]
```

# Creating a CNN Model
    it has hidden layers and M ax poll layrers an FC layers in last and a softmax layer
    Convolutional layer with 30 feature maps of size 5×5.
Pooling layer taking the max over 2*2 patches.
Convolutional layer with 15 feature maps of size 3×3.
Pooling layer taking the max over 2*2 patches.
Dropout layer with a probability of 20%.
Flatten layer.
Fully connected layer with 128 neurons and rectifier activation.
Fully connected layer with 50 neurons and rectifier activation.
Output layer.
    

```python
def cnn_model():
    model = Sequential()
    model.add(Conv2D(30,(5,5),input_shape=(1,28,28),activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Conv2D(15,(3,3),activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(128,activation='relu'))
    model.add(Dense(50,activation='relu'))
    model.add(Dense(num_classes,activation='softmax'))
    # compile and optimizer is adms
    model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
    return model
```

```python
model = cnn_model()
model.fit(x_train,y_train, validation_data=(x_valid,y_valid),epochs=10,batch_size=200)
# final 
scores = model.evaluate(x_valid,y_valid,verbose=0)
print('Large CNN ERROR:%.2f%%'%(100-scores[1]*100))
```

```python
model1 = cnn_model()
model1.fit(x_train,y_train, validation_data=(x_valid,y_valid),epochs=40,batch_size=64)
# final 
scores1 = model.evaluate(x_valid,y_valid,verbose=0)
print('Large CNN ERROR:%.2f%%'%(100-scores[1]*100))
```
